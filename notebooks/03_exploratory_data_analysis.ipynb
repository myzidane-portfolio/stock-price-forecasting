## üìò Introduction  
In this notebook, we perform **Exploratory Data Analysis (EDA)** to visualize and understand patterns in the AAPL stock data.

## ‚öôÔ∏è Import Libraries and Load Clean Data

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

ticker = "AAPL"
data = yf.download(ticker, start="2018-01-01", end="2025-01-01")
```

## üìà Plot Closing Price Over Time  
Visualize how the stock price changes through time.

```python
plt.figure(figsize=(12,6))
plt.plot(data["Close"], label="Closing Price", color="blue")
plt.title("AAPL Closing Price (2018‚Äì2025)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

## üìä Moving Averages  
Plot short-term and long-term moving averages to observe trend smoothing.

```python
data["MA50"] = data["Close"].rolling(window=50).mean()
data["MA200"] = data["Close"].rolling(window=200).mean()

plt.figure(figsize=(12,6))
plt.plot(data["Close"], label="Close")
plt.plot(data["MA50"], label="50-Day MA", color="orange")
plt.plot(data["MA200"], label="200-Day MA", color="green")
plt.title("AAPL Moving Averages")
plt.legend()
plt.show()
```

## üßæ Summary  
- Strong long-term growth trend observed.  
- Moving averages indicate trend shifts and momentum changes.
